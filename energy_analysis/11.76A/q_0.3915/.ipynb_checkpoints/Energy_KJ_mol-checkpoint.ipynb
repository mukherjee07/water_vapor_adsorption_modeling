```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Conversion factors
joule_to_kilojoules = 1e3
#KJ_to_KJ_mol = 6.022e23
conversion_factor = joule_to_kilojoules * (1.66054e23)  # Convert energy to Joules
mass_of_water_molecule = 18.015 / 1000  # kg/mol

# Get the list of files in the current directory
files = os.listdir('.')

# Separate files into two categories based on their names
host_ads_files = sorted(
    [f for f in files if f.endswith('_E.csv')],
    key=lambda x: ['AA', 'AR', 'AAR', 'Inner', 'Outer'].index(x.split('_')[0])
)
adsorption_files = sorted(
    [f for f in files if f.endswith('_Ads.csv')],
    key=lambda x: ['AA', 'AR', 'AAR', 'Inner', 'Outer'].index(x.split('_')[0])
)

# Define color scheme
folders = ['AA', 'AR', 'AAR', 'Inner', 'Outer']
colors = ['orange', 'cyan', 'green', 'red', 'silver']
color_map = dict(zip(folders, colors))
msize = [27.5, 25, 20, 15, 10]

# Function to calculate and plot energy per mole
def plot_energy_per_mole(energy_files, adsorption_files, title, ylabel, fontsize=12, ticksize=10):
    plt.figure(figsize=(15, 10), dpi=800)
    for i, (energy_file, adsorption_file) in enumerate(zip(energy_files, adsorption_files)):
        # Read energy and adsorption data
        energy_df = pd.read_csv(energy_file)
        adsorption_df = pd.read_csv(adsorption_file)

        # Extract data
        pressure = energy_df['pressure']  # Pressure in Pa
        vdw_energy = energy_df['vdw_energy'] * conversion_factor  # Convert energy to Joules
        adsorption_data = adsorption_df.iloc[:,1]  # Assuming adsorption data is in the second column

        # Constants for the system
        num_carbons = 960  # Number of carbon atoms in the system
        water_molecules = adsorption_data  # Sum of water molecules
        total_mass_water = water_molecules * mass_of_water_molecule  # Total mass of water in kg

        # Calculate moles
        moles_carbons = num_carbons / 12.01/1000  # Moles of carbon atoms
        moles_carbon = np.full(len(water_molecules),moles_carbons)
        moles_water = total_mass_water / mass_of_water_molecule  # Moles of water
        total_moles = moles_carbon + moles_water  # Total moles in the system

        # Calculate energy per mole
        energy_per_mole = vdw_energy / total_moles

        # Determine color based on file name prefix
        prefix = energy_file.split('_')[0]
        color = color_map.get(prefix, 'k')  # Default to black if prefix not found

        # Plot energy per mole
        plt.plot(pressure, energy_per_mole, marker='o', markersize=msize[i], linewidth=5,
                 markeredgewidth=2, markerfacecolor=color, linestyle=' ', label=f'{prefix}', zorder=3)

    plt.xscale('log')
    plt.xlabel('Pressure (Pa) $\\longrightarrow$', fontsize=fontsize)
    plt.ylabel(ylabel, fontsize=fontsize)
    plt.xticks(fontsize=ticksize)
    plt.yticks(fontsize=ticksize)
    plt.legend(loc="center left", prop={'size': 14})
    plt.grid(True, axis='both', which='minor', linestyle='--', alpha=0.5, linewidth=1.5, zorder=1)
    plt.grid(True, axis='both', which='major', linestyle='--', alpha=0.25, linewidth=1, zorder=1)
    plt.show()

# Plot energy per mole
plot_energy_per_mole(host_ads_files, adsorption_files, 'Energy Per Mole vs Pressure', '$E_{adsorption}$ (kJ/mol)', fontsize=14, ticksize=12)
```
